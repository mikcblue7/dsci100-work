```r
options(repr.matrix.max.rows = 20)
fast_food <- read_csv("data/fast_food.csv")
top_restaurants <- fast_food |>
    group_by(name) |>
    summarize(n = n()) |>
    arrange(desc(n)) |>
    slice_max(order_by = n, n=20)
#top_restaurants
population <- c(39.35, 12.72, 6.04, 28.64, 21.22, 10.52, 2.91, 19.51, 3.15)
subway_data <- fast_food |>
    filter(name == "Subway") |>
    group_by(st) |>
    summarize(n = n()) |>
    arrange(desc(n)) |>
    slice_max(order_by = n, n=9)
   
subway_per_cap <- cbind(subway_data, population) |>
    mutate(sub_per_capita = n/population) |>
    arrange(desc(sub_per_capita)) |>
    select(st, sub_per_capita)
#ubway_per_cap
subway_per_cap_plot <- subway_per_cap |> 
    ggplot(aes(x = fct_reorder(st, sub_per_capita, .desc = TRUE), y = sub_per_capita, 
               fill = fct_reorder(st, sub_per_capita)))+
        geom_bar(stat = "identity")+
        labs(x = "State", y = "Subway's per 1 million people", fill = "State") +
        theme(text = element_text(size = 20)) +
        ggtitle("Top 9 states with the highest ratio of Subway \n restaurants per 1 million people") +
        scale_fill_brewer(palette = "YlGn")
        
subway_per_cap_plot  
```
